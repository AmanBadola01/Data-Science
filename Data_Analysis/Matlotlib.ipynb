```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5]
y=[2,4,6,8,11]

#Crating a Line Plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic Line Plot')
```

```python
x= [1,2,3,4,5]
y=[2,4,6,8,11]

#Creating a cutomized plot
plt.plot(x,y, color= 'red', linestyle='--', marker="*")
plt.grid(True)
```

```python
#SubPlot

x= [1,2,3,4,5]
y1= [2,3,5,6,9]
y2= [3,6,9,12,15]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x,y1, color='green')
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(y1,x, color='red')
plt.title("Plot 2")
```

```python
#Bar Plot
categories=["A", "B", "C", "D"]
values = [10,15,20,25]

#Creating a bar plot
plt.bar(categories, values, color='purple')
plt.xlabel("categories")
plt.ylabel("values")
plt.title("Bar Graph")
```

```python
#Histogram
data= [1,1,1,2,3,4,5,5,5,3,2,3,1,2,3,5]
plt.hist(data, bins=4, color='yellow', edgecolor='black')
```

```python
#Scatter Plot

#Creating a Scatter Plot
x= [1,2,3,4,5]
y=[2,3,4,7,9]

plt.scatter(x,y, color='blue', marker='o')
```

```python
#Pie Chart
labels= ["A", "B", "C", "D"]
sizes= [30,40,50,60]
# explode = (0.2,0,0,0) #Moves out the first slice
explode=(0, 0.2,0,0)  #Moves out the second slice

#Create a pie Chart
plt.pie(sizes,explode=explode, labels= labels, autopct="%1.1f%%", shadow=True)
```
